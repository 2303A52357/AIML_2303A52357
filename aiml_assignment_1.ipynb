Part-02: Lab Task is to visualize the datasets or data using Matplotlib builtin function Graphs

Generate Scatterplot

Let's suppose we collected age (in years), height (in cm), and weight (in kilograms) information from few patients visitng a clinic.

```python
people=['Raju','Rahul','Ram','Sai','Sham']
age=[20,18,21,31,41]
weight=[50,50,100,100,109]
height=[140,150,160,170,180]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(age, weight)
plt.title("Relationship between Age and Weight")
plt.ylabel("Weight in Kilograms")
plt.xlabel("Age in Years")
plt.show()
```

**Generate Bar Plot**

```python
plt.figure(figsize=(12,4))
plt.title("people's weight in kgs",fontsize=16,fontstyle='italic')
#Main plot Function 'bar'
plt.bar(x=people,height=weight,width=0.6,color='orange',edgecolor='k',alpha=0.6)
plt.xlabel("People",fontsize=15)
plt.xticks(fontsize=12,rotation=30)
plt.yticks(fontsize=12)
plt.ylabel("Weight(in kgs)", fontsize=15)
plt.show()
```

**Generate Histogram**

```python
import numpy as np
plt.figure(figsize=(7,5))
# Main plot function 'hist'
plt.hist(age,color='yellow',edgecolor='b',alpha=0.75,bins=5)
plt.title("Histogram of patient Age",fontsize=18)
plt.xlabel('Age in Years',fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=15)
plt.show()
```
